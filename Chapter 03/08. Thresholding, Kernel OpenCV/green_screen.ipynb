```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(15, 15))
    plt.imshow(image)
```
